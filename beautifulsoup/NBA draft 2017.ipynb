```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/2017_NBA_draft")
page
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
body_content = soup.find('div',id = 'bodyContent')
```

```python
table = body_content.find_all('table')[3]
```

```python
row_name  = table.find_all(class_='sortkey') #names
```

```python
names = []
for row in row_name:
    names.append(row.text)
```

```python
rows = table.find_all('tr')[1:]
```

```python
rounds = []
picks = []
nationalities = []
teams = []
colleges = []
for row in rows:
    cols = row.find_all('td')
    rounds.append (cols[0].text)
    picks.append (cols[1].text)
    nationalities.append(cols[4].text)
    teams.append(cols[5].text)
    colleges.append(cols[6].text)
```

```python
import pandas as pd
```

```python
raw_data = {'Round':rounds,
           'College':colleges,
          'Players':names,
           'Team': teams,
           'Nationality': nationalities}

test = pd.DataFrame(raw_data,index= picks)
test.columns.name = "Selection"
```
